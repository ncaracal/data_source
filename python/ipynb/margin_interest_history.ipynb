# Binance Cross-Margin: API Interest-Rate History

Written by `data_source -d marginInterestRate`.

Path: `$TRADE_DATA/binance/spot/marginInterestRate/{ASSET}/{ASSET}_marginInterestRate_all.parquet`

Columns: `asset, timestamp, datetime, daily_interest_rate, vip_level`

```python
import os
from pathlib import Path
import pandas as pd

pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 160)

# --- config ---
ASSET = "STORJ"   # STO, STORJ, ...
N = 30            # number of latest rows to show

TRADE_DATA = os.environ.get("TRADE_DATA", "/ndata/trade/data")
base = Path(TRADE_DATA) / "binance" / "spot" / "marginInterestRate"
print("Base:", base)
print("Available assets:", sorted(p.name for p in base.iterdir() if p.is_dir()))
```

```python
path = base / ASSET / f"{ASSET}_marginInterestRate_all.parquet"
df = pd.read_parquet(path)
df["datetime"] = pd.to_datetime(df["timestamp"], unit="ms")
df = df.sort_values("timestamp").reset_index(drop=True)
print(f"{ASSET}: {len(df):,} rows, {df['datetime'].min()} .. {df['datetime'].max()}")
```
